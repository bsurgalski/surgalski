b) Use the FacetGrid from seaborn to create line plots to represent house_price_index for each state. Please make the line color red, set ticks on x axis for years ['1980', '2000', '2020'], which you may have to convert date type again, put those years as labels, and set the titles for each small plot to their state names. In addition, set the y label to 'house price index' and remove the x label since it's obvious that these are years.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
house_prices = pd.read_csv('/content/drive/MyDrive/CSC302/DATA/house_prices.csv')

# Filter for the selected states
states = ['Michigan', 'California', 'Texas', 'Indiana']
selected_states = house_prices[house_prices['state'].isin(states)]

# Convert the 'date' column to datetime format using .loc to avoid warning
selected_states.loc[:, 'date'] = pd.to_datetime(selected_states['date'])

# Create the FacetGrid for the line plot
g = sns.FacetGrid(selected_states, col="state", col_wrap=2, height=4)

# Map the lineplot onto the grid
g.map(sns.lineplot, 'date', 'house_price_index', color='red')

# Customize x-axis ticks and labels
g.set(xlim=(pd.to_datetime('1979-01-01'), pd.to_datetime('2021-01-01')))
for ax in g.axes.flat:  # Loop through all subplots (axes)
    ax.set_xticks([pd.to_datetime('1980-01-01'), pd.to_datetime('2000-01-01'), pd.to_datetime('2020-01-01')])
    ax.set_xticklabels(['1980', '2000', '2020'])

# Set y-label and remove x-label
g.set_axis_labels("", "House Price Index")

# Set titles for each facet (this is automatically handled by state)
g.set_titles("{col_name}")

# Adjust layout and show plot
plt.tight_layout()
plt.show()

#My first attempt at this had a warning. ChatGPT corrected this by changing the code for converting the date.


```

c) Draw a lineplot for 'house price perc', but color them based on the 'state' name which will have four lines in a single plot. Set a legend which will be at the bottom of your graph and have them horizantally positioned next to eachother. Again, set the y label to house_price_perc and remove the x label since it's obvious. (Note that my y label is still showing index. Please ignore that.)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
house_prices = pd.read_csv('/content/drive/MyDrive/CSC302/DATA/house_prices.csv')

# Convert 'date'
house_prices['date'] = pd.to_datetime(house_prices['date'])

# Filter the dataset to include only the selected states
selected_states = ['Michigan', 'California', 'Texas', 'Indiana']
house_prices_selected = house[house['state'].isin(selected_states)]

# Create a lineplot for 'house_price_perc' with color by 'state'
plt.figure(figsize=(10, 6))
sns.lineplot(data=house_prices_selected, x='date', y='house_price_perc', hue='state', palette='tab10')

# Set y-label and remove x-label
plt.ylabel('House Price Percentage')
plt.xlabel('')

# Add a legend at the bottom
plt.legend(title='State', loc='upper center', bbox_to_anchor=(0.5, -0.05), ncol=4)

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```
